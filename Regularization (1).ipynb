# **Splitting the data**

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.30,random_state=1)
```

# **Simple Linear Regression**

```python
le = LinearRegression()
le.fit(xtrain,ytrain)
ypred = le.predict(xtest)
le.score(xtrain,ytrain),le.score(xtest,ytest) #overfitted model
```

# **Lasso**

```python
l1 = Lasso() #by default leaning rate alpha = 1
l1.fit(xtrain,ytrain)
ypred = l1.predict(xtest)
le.score(xtrain,ytrain),le.score(xtest,ytest) #overfitted model
```

# **Tunning in Lasso**

```python
for i in range(100,200):
  l1 = Lasso(alpha = i)
  l1.fit(xtrain,ytrain)
  l1.predict(xtest)
  print(f"alpha : {i}, train : {l1.score(xtrain,ytrain)}, test : {l1.score(xtest,ytest)}")
```

```python
#at alpha = 175 we get both the accuracy similer
```

# **Final Model**

```python
le = Lasso(alpha = 175)
le.fit(xtrain,ytrain)
ypred = le.predict(xtest)
le.score(xtrain,ytrain),le.score(xtest,ytest)
```

# **Ridge**

```python
l2 = Ridge()
l2.fit(xtrain,ytrain)
ypred = l2.predict(xtest)
l2.score(xtrain,ytrain),l2.score(xtest,ytest) #overfitted model
```

# **Hyperparameter in Ridge**

```python
for i in range(1,50):
  l2 = Ridge(alpha = i)
  l2.fit(xtrain,ytrain)
  ypred = l2.predict(xtest)
  print(f"alpha : {i}, train : {l2.score(xtrain,ytrain)}, test : {l2.score(xtest,ytest)}")
```

```python
#at alpha = 7 we get both the accuracy similar
```

# **Final Model**

```python
l2 = Ridge(alpha = 7)
l2.fit(xtrain,ytrain)
ypred = l2.predict(xtest)
l2.score(xtrain,ytrain),l2.score(xtest,ytest)
```

# **3.ElasticNet**

```python
en = ElasticNet(alpha = 1)
en.fit(xtrain,ytrain)
ypred = en.predict(xtest)
en.score(xtrain,ytrain),en.score(xtest,ytest) #generalized model
```

# Tunning

```python
for i in range(1,10):
  en = ElasticNet(alpha = i)
  en.fit(xtrain,ytrain)
  ypred = en.predict(xtest)
  print(f"alpha : {i}, train : {en.score(xtrain,ytrain)}, test : {en.score(xtest,ytest)}")
```
