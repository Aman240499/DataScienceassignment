```python
import pandas as pd
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import plotly.express as pe
from bioinfokit.visuz import cluster
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/aishwaryamate/Machine-Learning/main/Dimensionality%20Reduction/t-sne.csv",index_col = 0)
df
```

```python
sns.countplot(x = df["cluster"])
```

```python
sns.pairplot(df)
plt.tight_layout
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
pca = PCA(n_components = 2)
pcs = pca.fit_transform(x)
pcs
```

```python
pe.scatter(x = pcs[:,0],y = pcs[:,1],color = y)
```

```python
#tSNE PLOT
```

```python
tsne = TSNE(n_components = 2)
tc = tsne.fit_transform(x)
tc
```

```python
pe.scatter(x = tc[:,0],y = tc[:,1],color = y)
```

# **Diabetes Dataset**

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/diabetes.csv')
df
```

```python
df.describe()
```

```python
sns.countplot(x = df["Outcome"])
```

```python
sns.pairplot(df)
plt.tight_layout
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#pca
pca = PCA(n_components = 2)
pcs =pca.fit_transform(x)
pcs
```

```python
pe.scatter(pcs[:,0],pcs[:,1],color = y)
```

```python
#tsne
tsne = TSNE(n_components = 2)
tc = tsne.fit_transform(x)
tc
```

```python
pe.scatter(x = tc[:,0],y = tc[:,1],color = y)
```

```python
#for image
cluster.tsneplot(tc, colorlist=y)
```
