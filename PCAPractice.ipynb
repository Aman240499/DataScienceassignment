```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/aishwaryamate/Machine-Learning/main/Dimensionality%20Reduction/PCA.csv")
df
```

```python
df.drop(columns = ["Univ"],inplace = True)
```

```python
df
```

```python
#1.standaidization of data
sc = StandardScaler()
x = sc.fit_transform(df)
x
```

```python
#coveriance of transproce version of data
cov = np.cov(x.T)
cov
```

```python
#eigen value and eigen vector values
values,vectors = np.linalg.eig(cov)
```

```python
values
```

```python
vectors
```

```python
#selecting highest eigen value and multiply x with that eigen vector value
pc1 = x@vectors[:,0]
pc1
```

```python
pc2 = x@vectors[:,1]
pc2
```

# **direct approch**

```python
pca = PCA(n_components=2)
pca.fit_transform(x)
```

# **full approch**

```python
pca = PCA()
pcs = pd.DataFrame(pca.fit_transform(x))
pcs
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
#scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number Of Dimensions")
plt.ylabel("Variance Ratio")
```

```python
#final model
pca = PCA(n_components = 2)
pca.fit_transform(x)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Wine.csv')
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,:-1]
y = df["Customer_Segment"]
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.20,random_state = 42)
```

```python
lg = LogisticRegression()
lg.fit(xtrain,ytrain)
ypred = lg.predict(xtest)
ypred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```
